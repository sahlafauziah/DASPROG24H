```python
mahasiswa = []

def menambahMahasiswa():
    try:
        nim = input("Masukkan NIM Harus 11 digit: ")
        if len(nim) != 11 or not nim.isdigit():
            raise ValueError("NIM harus terdiri dari 11 digit.")

        nama = input("Masukkan Nama (Min.6): ")
        if len(nama) < 6:
            raise ValueError("Nama harus minimal 6 karakter.")

        alamat = input("Masukkan Alamat (min.10) : ")
        if len(alamat) < 10:
            raise ValueError("Alamat harus minimal 10 karakter.")

        kelas = input("Masukkan Kelas (TI24H, TI24A, TI24G, TI24F): ").upper()
        if kelas not in ["TI24H", "TI24A", "TI24G", "TI24F"]:
            raise ValueError("Kelas hanya boleh TI24H, TI24A, TI24G, atau TI24F.")

        mahasiswa.append({
            "NIM": nim,
            "Nama": nama,
            "Alamat": alamat,
            "Kelas": kelas
        })

        print("Data mahasiswa berhasil ditambahkan.\n")

    except ValueError as e:
        print("Terjadi kesalahan" , e ,"\n")

def menampilkanMahasiswa():
    if not mahasiswa:
        print("Belum ada data mahasiswa.\n")
    else:
        print("\nDaftar Mahasiswa:")
        for index, data in enumerate(mahasiswa, start=1):
            print(f"{index}. NIM: {data['NIM']}, Nama: {data['Nama']}, Alamat: {data['Alamat']}, Kelas: {data['Kelas']}")
        print()

def menu():
    while True:
        print("=== MENU MAHASISWA ===")
        print("1. Tambah Data Mahasiswa")
        print("2. Tampilkan Data Mahasiswa")
        print("3. Keluar")
        pilihan = input("Pilih menu: ")

        try:
            if pilihan == "1":
                menambahMahasiswa()
            elif pilihan == "2":
                menampilkanMahasiswa()
            elif pilihan == "3":
                print("Program selesai.")
                break
            else:
                raise ValueError("Pilihan tidak valid. Harus 1, 2, atau 3.")
        except ValueError as error:
            print("Kesalahan:", error, "\n")

menu()
```
