```python
#Bulan - pengeluaran per minggu

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

bulan = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Des']

Minggu1 = np.random.randint(200000,500000,12)
Minggu2 = np.random.randint(200000,500000,12)
Minggu3 = np.random.randint(200000,500000,12)
Minggu4 = np.random.randint(200000,500000,12)

data = {
    'Bulan' : bulan,
    'Minggu_1' : Minggu1,
    'Minggu_2' : Minggu2,
    'Minggu_3' : Minggu3,
    'Minggu_4' : Minggu4,
}

df = pd.DataFrame(data)
df ['Total_pengeluaran'] = df[['Minggu_1', 'Minggu_2', 'Minggu_3', 'Minggu_4']].sum(axis=1)
df.describe()
```

```python
#Diagram Bar/Batang

plt.figure(figsize=(10,6)) #width,Height -> inch
plt.bar(df['Bulan'], df['Total_pengeluaran'],color='red')
plt.title("Pengeluaran Per Bulan")
plt.xticks(rotation=45)
plt.ylabel("Rupiah")
plt.xlabel("Bulan")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))

for minggu in ['Minggu_1', 'Minggu_2', 'Minggu_3', 'Minggu_4']:
    plt.plot(df['Bulan'], df[minggu], marker='o',label=minggu)

plt.title('Pengeluaran Per Minggu per Bulan')
plt.xlabel('Bulan')
plt.ylabel("Rupiah")
plt.legend()
plt.grid(True)
plt.xticks(rotation=-45)
plt.show()
```

```python
#Diagram pie Chart -> proporsi/persentase

plt.figure(figsize=(8,8))
plt.title('Proporsi Pengeluaran Per Bulan')
plt.pie(df['Total_pengeluaran'], labels=df['Bulan'], startangle=90, autopct="%1.1f%%")
plt.axis('equal')
plt.show()
```

```python
#AI , ML -> scikit ->Algoritma

from sklearn.linear_model import LinearRegression
df['Bulan_angka'] = np.arange(1,13)

x = df[['Bulan_angka']]
y = df['Total_pengeluaran']

model = LinearRegression()
model.fit(x,y)

prediksi = model.predict([[13]]) [0]
print(f"Pengeluaran di Bulan 13 : {prediksi:,.0f}")
```

```python
import requests
import pandas as pd
from io import StringIO

url = "https://data.slemankab.go.id/data/dataset/5fec3df5-159d-4ccc-8a64-54d1f21eaf49/resource/4de90d1e-732d-4a24-818c-d4b1d5403e1e/download/rekapitulasi-data-umkm-kabupaten-sleman-berdasarkan-skala-usaha-per-sektor-usaha-tahun-2023.csv"

# 1. Kirim permintaan dengan header 'User-Agent'
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}
resp = requests.get(url, headers=headers)
resp.raise_for_status()  # pastikan HTTP 200

# 2. Parsing CSV dari teks
csv_text = resp.text
df = pd.read_csv(StringIO(csv_text))

# 3. Cetak ringkas
print(df.head())          # 5 baris pertama
print(df.info())          # tipe dan jumlah non-null
print(df.describe())      # statistik kolom numerik
```
