```python
import numpy as np
import pandas as pd
n = 1000
usia = np.random.randint(20, 45, size=n)
gender = np.random.choice(['L', 'P'],size=n)
pendidikan = np.random.choice(['SMA','Diploma','Sarjana','Magister'], size=n)
pengalaman = np.random.poisson(3, size=n)
tes_tulis = np.random.normal(70, 10, size=n).clip(0, 100)
tes_wawancara = np.random.normal(75, 8, size=n).clip(0, 100)
keahlian = np.random.choice(['Dasar','Menengah','Lanjutan'], size=n)
kesesuaian = np.random.choice(['Rendah','Sedang','Tinggi'], size=n)
minat = np.random.choice(['Rendah','Sedang','Tinggi'], size=n)
diterima = np.random.choice(['Ya','Tidak','Mungkin'],size=n)

df = pd.DataFrame({
    'usia' : usia,
    'gender' : gender,
    'pendidikan': pendidikan,
    'pengalaman' : pengalaman,
    'tes_tulis' : tes_tulis,
    'tes_wawancara' : tes_wawancara,
    'keahlian' : keahlian,
    'kesesuaian' : kesesuaian,
    'minat' : minat,
    'diterima' : diterima
})

print(df)
print(df.info()) 
print(df.head())

df.to_csv("Data Tes Interview.csv")
    
```
