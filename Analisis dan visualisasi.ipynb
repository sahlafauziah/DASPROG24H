```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("netflix.csv")
df
```

```python
#Analisi type Movie dan TV Show

type_counts = df['type'].value_counts()

plt.bar(type_counts.index, type_counts.values, color = ['skyblue', 'salmon'])
plt.title('Distribusi Tipe konten')
plt.xlabel('Tipe')
plt.ylabel('Jumlah')
plt.grid(axis='y')
plt.show()
```

```python
#Analisis Genre terpopuler

genre_series = df['listed_in'].dropna().str.split('.')
all_genres = genre_series.explode()
genre_counts = all_genres.value_counts().head(10)

plt.barh(genre_counts.index[::-1], genre_counts.values[::-1], color = 'pink')
plt.title('10 Genre Terpopuler')
plt.xlabel('Jumlah')
plt.ylabel('Genre')
plt.grid(axis='x')
plt.show()
```

```python
#Analisis Tren Konten pertahun 

year_counts = df['release_year'].value_counts().sort_index()

plt.plot(year_counts.index, year_counts.values, color='yellow', marker= 'o')
plt.title('Jumlah Konten Dirilis PerTahun')
plt.xlabel('Tahun')
plt.ylabel('Jumlah')
plt.grid(True)
plt.show()
```

```python
#Analisis Negara Asal Terbanyak

country_series = df['country'].dropna().str.split('.')
all_countries = country_series.explode()
country_counts = all_countries.value_counts().head(10)

plt.barh(country_counts.index[::-1], country_counts.values[::-1], color = 'red')
plt.title('10 Negara Asal Konten Terbanyak')
plt.xlabel('Jumlah')
plt.ylabel('Negara')
plt.grid(axis='x')
plt.show()
```

```python
#Analisis Durasi Film

movie_df = df[df['type'] == 'Movie'].copy()
movie_df['duration_minutes'] = movie_df['duration'].str.extract('(\d+)').astype(float)

plt.hist(movie_df['duration_minutes'].dropna(), bins=30, color = 'darkred', edgecolor='black')
plt.title('Distribusi Durasi Film(Menit)')
plt.xlabel('Durasi')
plt.ylabel('Jumlah Film')
plt.grid(True)
plt.show()
```
